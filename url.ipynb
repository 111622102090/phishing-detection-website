```python
import re
from urllib.parse import urlparse

def extract_features(url):
    features = []

    # Feature 1: Length of the URL
    features.append(len(url))

    # Feature 2: Count of digits in the URL
    features.append(sum(c.isdigit() for c in url))

    # Feature 3: Count of special characters in the URL
    features.append(len(re.findall(r'[\W_]', url)))

    # Feature 4: Count of uppercase letters in the URL
    features.append(sum(1 for c in url if c.isupper()))

    # Feature 5: Count of lowercase letters in the URL
    features.append(sum(1 for c in url if c.islower()))

    # Feature 6: Number of dots (.) in the domain
    domain = urlparse(url).netloc
    features.append(domain.count('.'))

    # Feature 7: Number of slashes (/) in the URL
    features.append(url.count('/'))

    # Feature 8: Length of the domain name
    features.append(len(domain))

    # Feature 9: Count of hyphens (-) in the domain
    features.append(domain.count('-'))

    # Feature 10: Count of query parameters (?)
    features.append(url.count('?'))

    # Feature 11: Check if the URL uses HTTPS
    features.append(int(url.lower().startswith('https://')))

    # Feature 12: Count of subdomains in the URL
    subdomains = domain.split('.')
    features.append(len(subdomains) - 2)  # Subdomains count (subtracting domain and TLD)

    # Feature 13: Count of numeric subdomains
    numeric_subdomains = sum(1 for sub in subdomains[:-2] if sub.isdigit())
    features.append(numeric_subdomains)

    # Feature 14: Length of the path (after the domain)
    features.append(len(urlparse(url).path))

    # Feature 15: Check if there is an IP address in the URL
    features.append(int(bool(re.match(r'\d+\.\d+\.\d+\.\d+', domain))))

    # Feature 16: Check if the URL contains 'login' (indicating phishing behavior)
    features.append(int('login' in url.lower()))

    # Feature 17: Check if the URL contains 'secure' (indicating phishing behavior)
    features.append(int('secure' in url.lower()))

    return features
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import re
from urllib.parse import urlparse



# Sample data with URLs and labels (replace this with actual data)
data = [
    {"url": "http://example.com", "label": 0},
    {"url": "https://secure-login.com", "label": 1},
    {"url": "https://www.amazon.com", "label": 0},
    # Add more URLs and labels here
]

# Extract features and labels
urls = [item['url'] for item in data]
labels = [item['label'] for item in data]
features = [extract_features(url) for url in urls]

# Convert to DataFrame for easy manipulation
df = pd.DataFrame(features)
df['Label'] = labels  # Add the target variable (label)

# Separate features and target
X = df.drop('Label', axis=1)
y = df['Label']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train a RandomForestClassifier
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = rf_classifier.predict(X_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Display the classification report (precision, recall, f1-score)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
import joblib

# Save the RandomForest model
joblib.dump(rf_classifier, 'rf_classifier.pkl')

# Save the StandardScaler
joblib.dump(scaler, 'scaler.pkl')

```

```python
# Load the saved RandomForest model
rf_classifier = joblib.load('rf_classifier.pkl')

# Load the saved StandardScaler
scaler = joblib.load('scaler.pkl')
```

```python
def predict_url(url):
    # Extract features from the URL
    new_var = extract_features(url)
    features = new_var

    # Scale the features using the saved scaler
    scaled_features = scaler.transform([features])

    # Make the prediction using the trained model
    prediction = rf_classifier.predict(scaled_features)

    # Return the result: "Phishing" or "Legitimate"
    return "Phishing" if prediction == 1 else "Legitimate"

# Test the function with a new URL
url_to_check = "http://secure-login1234.com"  # Replace with any URL you want to check
result = predict_url(url_to_check)
print(f"The URL '{url_to_check}' is {result}.")

```
